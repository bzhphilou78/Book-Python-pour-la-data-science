# 6. Manipulation avancée des données avec Pandas
## 6.1.	Les opérations groupby
### 6.1.1.	groupby sur une colonne

```python
donnees.groupby("Sport").mean()
```

```python
donnees.groupby("Sport").mean().loc[:,['Age','Height','Weight']]
```

### 6.1.2. groupby sur plusieurs colonnes

```python
donnees.groupby(["Sport", "Sex"]).mean().loc[:,['Age','Height','Weight']]
```

### 6.1.3. Appliquer plusieurs fonctions avec la méthode groupby et la méthode aggregate

```python
donnees.groupby(['Sport', 'Sex']).agg(['median','mean']).loc[:,['Age','Height','Weight']]
```

```python
donnees.groupby(['Sport', 'Sex']).agg({'Age' : ['median','mean'], 'Height' : ['median', 'sum'], 'Weight' : ['mean','median']})
```

## 6.2.	Appliquer une fonction à un dataframe avec la méthode apply

```python
import numpy as np
print(donnees.loc[:,["Age", "Height", "Weight"]].apply(np.mean, axis=0))
```

```python
print(donnees[["Age", "Height", "Weight"]].apply(lambda x : x * 10))
```

```python
donnees['Height'] = donnees['Height'].apply(lambda x : x / 100)
donnees.head()
```

## 6.3.	Remodeler / ré-organiser des dataframes
### 6.3.1.	Pivotage : la méthode pivot_table

```python
pd.set_option('display.max_rows', 70)
print(donnees.pivot_table(index='Sport', columns='Medal', values="Sex", aggfunc='max'))
pd.set_option('display.max_rows', 10)
```

### 6.3.2. Les méthodes stack (empiler) et unstack (désempiler)

```python
multi_index=donnees.groupby(["Sport", "Sex"]).mean()[['Age','Height','Weight']]
print("Sans stack")
print(multi_index)
print("Après stack")
print(multi_index.stack())
```
